```python
import numpy as np 
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
dataset = pd.read_csv(r"C:\Users\gtanu\Downloads\archive (4)\Financial Distress.csv")
dataset
```

```python
dataset.describe()
```

```python
dataset['Company'].unique()
```

```python
x_train = dataset.iloc[:,3:].values
y_train = dataset.iloc[:, 2].values
y_test = dataset.iloc[:, 2].values
x_test = dataset.iloc[:,3:].values
x_train
```

```python
y_train_binary = (y_train <= -0.50).astype(int)
y_test_binary = (y_test <= -0.50).astype(int)
```

```python
my_obj = KNeighborsClassifier()
my_model = my_obj.fit(x_train, y_train_binary)
y_pred = my_model.predict(x_test)
```

```python
accuracy = metrics.accuracy_score(y_test_binary, y_pred, normalize=True)
accuracy
```

```python
cm = confusion_matrix(y_pred, y_test_binary)
cf = classification_report(y_pred, y_test_binary)
print(cm)
print(cf)
```

```python
y_binary = (y_train <= -0.50).astype(int)
```

```python
x_train, x_test, y_train_binary, y_test_binary = train_test_split(x_test, y_binary, test_size=0.3, random_state=10)
knn = KNeighborsClassifier()
knn.fit(x_train, y_train_binary)
y_pred = knn.predict(x_test)

print("Accuracy:", metrics.accuracy_score(y_test_binary, y_pred))
print("F1 Score:", metrics.f1_score(y_test_binary, y_pred))
```
